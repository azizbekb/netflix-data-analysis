# Netflix Data Analysis

##### Importing Libraries

```python
import numpy as np #linear algebra
import pandas as pd #data processing
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Loading the Dataset

```python
netflix=pd.read_csv("netflix_titles.csv")
netflix.head(10)
```

```python
df=netflix.copy()
df.shape
```

```python
df=df.dropna()
df.shape
```

```python
df.head(10)
```

##### Convert Date Time Format

```python
df["date_added"]=pd.to_datetime(df['date_added'].str.strip())
df['day_added']=df['date_added'].dt.day
df['year_added']=df['date_added'].dt.year
df['month_added']=df['date_added'].dt.month
df['year_added'].astype(int);
df['day_added'].astype(int);
```

```python
df.head()
```

## Data Visualization

```python
sns.countplot(x='type', hue='type', data=netflix, palette={'Movie': 'skyblue', 'TV Show': 'orange'}, legend=False)
fig=plt.gcf()
fig.set_size_inches(10,10)
plt.title('Type')
plt.show()
```

```python
unique_ratings = netflix['rating'].dropna().unique()
palette = sns.color_palette("Set2", n_colors=len(unique_ratings))
ax=sns.countplot(x='rating', data=netflix, hue='rating', palette=palette)
plt.xticks(rotation=90, ha="right")
fig = plt.gcf()
fig.set_size_inches(13,13)
plt.title('Rating')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x='rating', hue='type', data=netflix)
plt.title("Relation between Type and Rating")
plt.show()
```

```python
labels = ['Movie', 'TV show']
size = netflix['type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0, 1, 2))
explode = [0, 0.1]
plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size,labels=labels, colors = colors, explode = explode, shadow = True, startangle = 90)
plt.title('Distribution of Type', fontsize = 25)
plt.legend()
plt.show()
```

```python
netflix['rating'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.show()
```
